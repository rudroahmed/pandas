```python
import pandas as pd
```

```python
data = {'age':[25,26,27,28,29],
        'Salary':[30000,40000,50000,60000,70000]}
df = pd.DataFrame(data)
print(df)
```

```python
df.info()
```

```python
df.dtypes
```

```python
df['age']=df['age'].astype('int16')
df['Salary']=df['Salary'].astype('int16')
df
```
